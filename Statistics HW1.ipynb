5. Construct a histogram with 10 bins. Locate on your histogram 1, 2, 3
standard deviation far points from mean to both sides. Count the data points in each of 3 intervals and compare this with the empirical rule and interpret. (5 points)

```python
df
```

```python
means = df.mean()
std_devs = df.std()

std_1_intervals = []
std_2_intervals = []
std_3_intervals = []

for column in df.columns:
    mean = means[column]
    std_dev = std_devs[column]
    std_1_intervals.append(mean + std_dev)
    std_2_intervals.append(mean + 2 * std_dev)
    std_3_intervals.append(mean + 3 * std_dev)

plt.figure(figsize=(10, 6))
for column in df.columns:
    plt.hist(df[column], bins=10, alpha=0.7, label=column)

for std_1, std_2, std_3 in zip(std_1_intervals, std_2_intervals, std_3_intervals):
    plt.axvline(x=std_1, color='red', linestyle='--', label='1 Std Dev')
    plt.axvline(x=std_2, color='orange', linestyle='--', label='2 Std Dev')
    plt.axvline(x=std_3, color='green', linestyle='--', label='3 Std Dev')

plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

count_std_1 = df.apply(lambda col: ((col >= means[col.name] - std_devs[col.name]) & (col <= means[col.name] + std_devs[col.name])).sum())
count_std_2 = df.apply(lambda col: ((col >= means[col.name] - 2 * std_devs[col.name]) & (col <= means[col.name] + 2 * std_devs[col.name])).sum())
count_std_3 = df.apply(lambda col: ((col >= means[col.name] - 3 * std_devs[col.name]) & (col <= means[col.name] + 3 * std_devs[col.name])).sum())

print("Data points in 1 Std Dev interval:")
print(count_std_1)
print("\nData points in 2 Std Dev interval:")
print(count_std_2)
print("\nData points in 3 Std Dev interval:")
print(count_std_3)
```

As we can see for the first interval there is a small difference between them. But for the 2nd and third Interval EU_Sales and Global_Sales are equal.
